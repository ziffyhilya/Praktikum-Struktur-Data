Bubble Sort

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        for j in range(n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

# Contoh penggunaan
data = [64, 34, 25, 12, 22, 11, 90]
sorted_data = bubble_sort(data.copy()) # Menggunakan copy agar data asli tidak berubah
print("Data sebelum sorting:", data)
print("Data setelah Bubble Sort:", sorted_data)
```

Implementasi Selection Sort

```python
def selection_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        min_index = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]
    return arr

# Contoh penggunaan
data = [64, 25, 12, 22, 11]
sorted_data = selection_sort(data.copy())
print("Data sebelum sorting:", data)
print("Data setelah Selection Sort:", sorted_data)
```

Implementasi Insertion Sort

```python
def insertion_sort(arr):
    n = len(arr)
    for i in range(1, n):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

# Contoh penggunaan
data = [64, 34, 25, 12, 22, 11, 90]
sorted_data = insertion_sort(data.copy())
print("Data sebelum sorting:", data)
print("Data setelah Insertion Sort:", sorted_data)
```

Implementasi Merge Sort

```python
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left_half = arr[:mid]
        right_half = arr[mid:]

        merge_sort(left_half)
        merge_sort(right_half)

        i = j = k = 0

        while i < len(left_half) and j < len(right_half):
          if left_half[i] < right_half[j]:
            arr[k] = left_half[i]
            i += 1
          else:
            arr[k] = right_half[j]
            j += 1
          k += 1

        while i < len(left_half):
          arr[k] = left_half[i]
          i += 1
          k += 1

        while j < len(right_half):
          arr[k] = right_half[j]
          j += 1
          k += 1
    return arr

# Contoh penggunaan
data = [64, 25, 12, 22, 11]
sorted_data = merge_sort(data.copy())
print("Data sebelum sorting:", data)
print("Data setelah Merge Sort:", sorted_data)
```

Implementasi Quick Sort

```python
def quick_sort(arr):
    if len(arr) <= 1:
      return arr

    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]

    return quick_sort(left) + middle + quick_sort(right)

# Contoh penggunaan
data = [64, 25, 12, 22, 11]
sorted_data = quick_sort(data.copy())
print("Data sebelum sorting:", data)
print("Data setelah Quick Sort:", sorted_data)
```

Tugas Algoritma Sorting
Sistem Manajemen Data Pengguna

```python
def selection_sort(arr, key):
    n = len(arr)
    for i in range(n - 1):
        min_index = i
        for j in range(i + 1, n):
            if arr[j][key] < arr[min_index][key]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]
    return arr

# Contoh data pelanggan aplikasi e-commerce
data_pelanggan = [
    {"nama": "KENJI", "umur": 20, "total_belanja": 290000},
    {"nama": "FAHMA", "umur": 19, "total_belanja": 170000},
    {"nama": "AQILA", "umur": 20, "total_belanja": 280000},
    {"nama": "RAPLY", "umur": 20, "total_belanja": 90000},
    {"nama": "NOPAL", "umur": 20, "total_belanja": 350000},
    {"nama": "UFII", "umur": 19, "total_belanja": 150000},
    {"nama": "EVI", "umur": 36, "total_belanja": 270000},
    {"nama": "SINDI", "umur":29, "total_belanja": 350000},
    {"nama": "OCTA", "umur": 27, "total_belanja": 260000},
    {"nama": "DAFFY", "umur": 17, "total_belanja": 120000}
]

# Sorting berdasarkan berbagai kriteria
sorted_nama = selection_sort(data_pelanggan.copy(), "nama")
sorted_umur = selection_sort(data_pelanggan.copy(), "umur")
sorted_belanja = selection_sort(data_pelanggan.copy(), "total_belanja")

# Output
print("Data sebelum sorting:")
for data in data_pelanggan:
    print(data)

print("\nData setelah sorting berdasarkan Nama (A-Z):")
for data in sorted_nama:
    print(data)

print("\nData setelah sorting berdasarkan Umur (dari yang termuda ke tertua):")
for data in sorted_umur:
    print(data)

print("\nData setelah sorting berdasarkan Total Belanja (dari yang terkecil ke terbesar):")
for data in sorted_belanja:
    print(data)
```
