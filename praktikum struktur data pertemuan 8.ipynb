Representasi Graf dengan Adjacency List

```python
# Membuat graf tak berarah menggunakan adjacency list
graph = {
    "A": ["B", "C"],
    "B": ["A", "D", "E"],
    "C": ["A", "F"],
    "D": ["B"],
    "E": ["B", "F"],
    "F": ["C", "E"]
}


# Fungsi untuk menampilkan graf
def print_graph(graph):
    for node, neighbors in graph.items():
        print(f"{node} -> {', '.join(neighbors)}")


# Menampilkan isi graf
print_graph(graph)
```

Traversal Graf menggunakan BFS


```python
from collections import deque


# Fungsi Breadth First Search (BFS)
def bfs(graph, start):
    visited = set()          # Menyimpan node yang sudah dikunjungi
    queue = deque([start])   # Queue untuk proses BFS

    while queue:
        vertex = queue.popleft()

        if vertex not in visited:
            print(vertex, end=" ")
            visited.add(vertex)

            # Menambahkan tetangga yang belum dikunjungi ke queue
            for neighbor in graph[vertex]:
                if neighbor not in visited:
                    queue.append(neighbor)


# Contoh graf
graph = {
    "A": ["B", "C"],
    "B": ["A", "D", "E"],
    "C": ["A", "F"],
    "D": ["B"],
    "E": ["B", "F"],
    "F": ["C", "E"]
}


# Contoh pemanggilan BFS
bfs(graph, "A")
```

Traversal Graf menggunakan DFS


```python
# Fungsi Depth First Search (DFS)
def dfs(graph, start, visited=None):
    # Membuat set visited jika belum ada
    if visited is None:
        visited = set()

    # Mengecek apakah node sudah dikunjungi
    if start not in visited:
        print(start, end=" ")
        visited.add(start)

        # Menelusuri setiap tetangga dari node
        for neighbor in graph[start]:
            dfs(graph, neighbor, visited)


# Contoh graf
graph = {
    "A": ["B", "C"],
    "B": ["A", "D", "E"],
    "C": ["A", "F"],
    "D": ["B"],
    "E": ["B", "F"],
    "F": ["C", "E"]
}


# Contoh pemanggilan DFS
dfs(graph, "A")
```

Tugas

```python
from collections import deque


# =========================================
# Membuat graf tak berarah
# =========================================
graph = {
    "A": ["B", "C", "D"],
    "B": ["A", "C", "E"],
    "C": ["A", "B", "D", "F"],
    "D": ["A", "C", "E"],
    "E": ["B", "D", "F"],
    "F": ["C", "E"]
}


# =========================================
# Fungsi untuk menampilkan graf
# =========================================
def print_graph(graph):
    print("Adjacency List Graph:")

    for node, neighbors in graph.items():
        print(f"{node} -> {', '.join(neighbors)}")


# =========================================
# BFS (mengembalikan list urutan kunjungan)
# =========================================
def bfs(graph, start):
    visited = set()
    queue = deque([start])
    result = []

    while queue:
        vertex = queue.popleft()

        if vertex not in visited:
            visited.add(vertex)
            result.append(vertex)

            for neighbor in graph[vertex]:
                if neighbor not in visited:
                    queue.append(neighbor)

    return result


# =========================================
# DFS
# =========================================
def dfs(graph, start, visited=None):
    if visited is None:
        visited = set()

    if start not in visited:
        print(start, end=" ")
        visited.add(start)

        for neighbor in graph[start]:
            dfs(graph, neighbor, visited)


# =========================================
# Mencari jalur menggunakan DFS
# =========================================
def find_path(graph, start, end, path=None):
    if path is None:
        path = []

    path = path + [start]

    # Jika simpul tujuan ditemukan
    if start == end:
        return path

    # Jika simpul tidak ada dalam graf
    if start not in graph:
        return None

    # Menelusuri tetangga
    for neighbor in graph[start]:
        if neighbor not in path:
            new_path = find_path(graph, neighbor, end, path)

            if new_path:
                return new_path

    return None


# =========================================
# Mengecek apakah graf terhubung
# =========================================
def is_connected(graph):
    start_node = next(iter(graph))  # Ambil simpul pertama
    visited = set()

    # DFS untuk menandai simpul yang dikunjungi
    def dfs_connected(node):
        visited.add(node)

        for neighbor in graph[node]:
            if neighbor not in visited:
                dfs_connected(neighbor)

    dfs_connected(start_node)

    # Jika semua simpul dikunjungi -> graf terhubung
    return len(visited) == len(graph)


# =========================================
# Menampilkan graf
# =========================================
print_graph(graph)


# =========================================
# Menjalankan BFS
# =========================================
print("\nHasil BFS dari simpul A:")
print(bfs(graph, "A"))


# =========================================
# Menjalankan DFS
# =========================================
print("\nHasil DFS dari simpul A:")
dfs(graph, "A")


# =========================================
# Mencari jalur
# =========================================
print("\n\nMencari jalur dari A ke F:")
print(find_path(graph, "A", "F"))


# =========================================
# Mengecek keterhubungan graf
# =========================================
print("\nApakah graf terhubung?")
print(is_connected(graph))
```
