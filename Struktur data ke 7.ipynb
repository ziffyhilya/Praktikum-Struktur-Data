Membuat Binary Tree Secara Manual

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

# Membuat tree secara manual
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)

# Fungsi inorder traversal
def inorder(node):
    if node:
        inorder(node.left)
        print(node.data, end=' ')
        inorder(node.right)

print("Hasil inorder traversal dari binary tree:")
inorder(root)
```

Membuat Binary Search Tree

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None


class BinarySearchTree:
    def __init__(self):
        self.root = None

    def insert(self, root, data):
        # Jika node kosong, buat node baru
        if root is None:
            return Node(data)

        # Jika data lebih kecil dari root
        if data < root.data:
            root.left = self.insert(root.left, data)

        # Jika data lebih besar atau sama dengan root
        else:
            root.right = self.insert(root.right, data)

        return root


# Membuat objek BST
bst = BinarySearchTree()
root = None

# Data yang akan dimasukkan
data_list = [50, 30, 70, 20, 40, 60, 80]

# Memasukkan data ke BST
for value in data_list:
    root = bst.insert(root, value)

print("Binary Search Tree berhasil dibuat.")
```

Traversal pada BST

```python
# Fungsi inorder traversal
def inorder(node):
    if node:
        inorder(node.left)
        print(node.data, end=' ')
        inorder(node.right)


# Fungsi preorder traversal
def preorder(node):
    if node:
        print(node.data, end=' ')
        preorder(node.left)
        preorder(node.right)


# Fungsi postorder traversal
def postorder(node):
    if node:
        postorder(node.left)
        postorder(node.right)
        print(node.data, end=' ')


# Menampilkan hasil traversal
print("Inorder Traversal:")
inorder(root)

print("\nPreorder Traversal:")
preorder(root)

print("\nPostorder Traversal:")
postorder(root)
```

Pencarian Nilai dalam BST

```python
# Fungsi pencarian pada Binary Search Tree
def search(node, key):
    # Jika node kosong atau data ditemukan
    if node is None or node.data == key:
        return node

    # Jika key lebih kecil, cari ke subtree kiri
    if key < node.data:
        return search(node.left, key)

    # Jika key lebih besar, cari ke subtree kanan
    return search(node.right, key)


# Uji pencarian data
key = 60
result = search(root, key)

if result:
    print(f"{key} ditemukan dalam tree.")
else:
    print(f"{key} tidak ditemukan.")


key = 25
result = search(root, key)

if result:
    print(f"{key} ditemukan dalam tree.")
else:
    print(f"{key} tidak ditemukan.")
```

```python
# Struktur Tree berdasarkan identitas saya

class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None


# =========================
# Data Identitas
# =========================

nama = "ZIFFY HILYA"
npm = "25903487"
tanggal_lahir = "24-04-2010"

# Menghitung jumlah huruf nama (tanpa spasi)
jumlah_huruf = len(nama.replace(" ", ""))

# Menghitung jumlah huruf vokal
vokal = "aiueoAIUEO"
jumlah_vokal = sum(1 for huruf in nama if huruf in vokal)

# Mengambil 2 digit terakhir NPM
dua_digit_npm = int(npm[-2:])

# Mengambil tanggal dan bulan lahir
tanggal = int(tanggal_lahir.split("-")[0])  # DD
bulan = int(tanggal_lahir.split("-")[1])    # MM


# =========================
# Membuat Binary Tree
# =========================

root = Node(jumlah_huruf)

# Anak kiri dan kanan root
root.left = Node(jumlah_vokal)
root.right = Node(dua_digit_npm)

# Level berikutnya
root.left.left = Node(tanggal)
root.right.right = Node(bulan)


# =========================
# Fungsi Inorder Traversal
# =========================

def inorder(node):
    if node:
        inorder(node.left)
        print(node.data, end=' ')
        inorder(node.right)


# =========================
# Menampilkan Hasil
# =========================

print("Hasil Inorder Traversal:")
inorder(root)
```

Tugas 2: Program Binary Search Tree (BST)

```python
# Program Binary Search Tree (BST)
# Dibuat untuk memenuhi tugas traversal dan pencarian data

# =========================
# Membuat Node BST
# =========================
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

# =========================
# Membuat class BST
# =========================
class BST:
    def __init__(self):
        self.root = None

    # Fungsi insert data ke BST
    def insert(self, data):
        if self.root is None:
            self.root = Node(data)
        else:
            self._insert(self.root, data)

    # Fungsi rekursif insert
    def _insert(self, current, data):
        if data < current.data:
            if current.left is None:
                current.left = Node(data)
            else:
                self._insert(current.left, data)
        elif data > current.data:
            if current.right is None:
                current.right = Node(data)
            else:
                self._insert(current.right, data)

    # =========================
    # Traversal Inorder
    # =========================
    def inorder(self, node):
        if node:
            self.inorder(node.left)
            print(node.data, end=" ")
            self.inorder(node.right)

    # =========================
    # Traversal Preorder
    # =========================
    def preorder(self, node):
        if node:
            print(node.data, end=" ")
            self.preorder(node.left)
            self.preorder(node.right)

    # =========================
    # Traversal Postorder
    # =========================
    def postorder(self, node):
        if node:
            self.postorder(node.left)
            self.postorder(node.right)
            print(node.data, end=" ")

    # =========================
    # Fungsi pencarian data
    # =========================
    def search(self, node, key):
        if node is None:
            return False

        if node.data == key:
            return True
        elif key < node.data:
            return self.search(node.left, key)
        else:
            return self.search(node.right, key)

# =========================
# Data angka unik
# Contoh data:
# 3 digit terakhir NPM = 245
# Tanggal lahir = 12/08 → 12 dan 8
# ASCII huruf depan nama = Z dan I
# Z = 90
# I = 73
# Tambahan angka bebas = 17 dan 99
# =========================
data = [245, 12, 8, 90, 73, 17, 99]

# Membuat objek BST
tree = BST()

# Memasukkan data ke BST
for angka in data:
    tree.insert(angka)

# =========================
# Menampilkan Traversal
# =========================
print("Data dalam BST:", data)

print("\nTraversal Inorder:")
tree.inorder(tree.root)

print("\n\nTraversal Preorder:")
tree.preorder(tree.root)

print("\n\nTraversal Postorder:")
tree.postorder(tree.root)

# =========================
# Fitur pencarian data
# =========================

# Dua digit terakhir NPM
cari1 = 45

# Angka yang tidak ada dalam list
cari2 = 100

print("\n\nHasil Pencarian:")

if tree.search(tree.root, cari1):
    print(f"Data {cari1} ditemukan dalam BST")
else:
    print(f"Data {cari1} tidak ditemukan dalam BST")

if tree.search(tree.root, cari2):
    print(f"Data {cari2} ditemukan dalam BST")
else:
    print(f"Data {cari2} tidak ditemukan dalam BST")
```
