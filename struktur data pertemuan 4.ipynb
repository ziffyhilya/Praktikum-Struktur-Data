Queue dengan List

```python
# program python untuk mendemontrasi antrian menggunakan using list
queue= []

# fungsi append() digunakan untuk menmbah elemen ke dalam queue
queue.append('m')
queue.append('o')
queue.append('h')
queue.append('a')
queue.append('n')

print('Initial queue')
print(queue)

#fungsi pop(0) untuk menghapus elemen dari queue dalam urutan FIFO
print('\nElement yang di hapus dari queue')
print(queue.pop(0))
print(queue.pop(0))
print(queue.pop(0))

print('\nQueue setelah menghapus elemen')
print(queue)

#uncommenting print(queue.pop()) will cause in IndexError as the queue is now emty
```

Queue dengan collections.deque

```python
# program python untuk mendemontrasi antrian menggunakan collections.deque
from collections import deque

queue= deque()

# fungsi append() digunakan untuk menmbah elemen ke dalam queue
queue.append('m')
queue.append('o')
queue.append('h')
queue.append('a')
queue.append('n')

print('Initial queue')
print(queue)

#fungsi popleft(0) untuk menghapus elemen dari queue dalam urutan FIFO
print('\nElement yang di hapus dari queue')
print(queue.popleft())
print(queue.popleft())
print(queue.popleft())

print('\nQueue setelah menghapus elemen')
print(queue)

#uncommenting print(queue.pop()) will cause in IndexError as the queue is now emty
```

Queue dengan queue.Queue

```python
# program python untuk mendemontrasi queue menggunakan queue.Queue
from queue import Queue

#menganalisasi queue
qu = Queue(maxsize=5)

#qsize() menampilkan jumlah elemen dalam queue
print(qu.qsize())

#fungsi put() untuk memasukan elemen kedalam queue
qu.put('z')
qu.put('i')
qu.put('f')
qu.put('f')
qu.put('y')

print("Full: ", qu.full())
print("Size: ", qu.qsize())

#fungsi get() untuk menghapus dequeue dari queue
print("\nElemen dequeue yang di hapus dari queue")
print(qu.get())
print(qu.get())
print(qu.get())

print("\nEmpty: ", qu.empty())
print("Full: ", qu.full())
```
