Stack dengan List

```python
#Mendemonstrasikan implementasi stack menggunakan list
stack = []

# fungsi append() untuk menambahkan elemen kedalam stack
stack.append('ziffy')
stack.append('mohan')
stack.append('baskara mahendra')
stack.append('killua')
stack.append('kadam sidik')

print('Initial stack')
print(stack)

# menghapus elemen terakhir dari stack dengan pop()
print('\nElements popped from stack:')
print(stack.pop())
print(stack.pop())
print(stack.pop())

print('\nStack after elements are popped:')
print(stack)

#uncommenting print(stack.pop()) will cause an IndexError as the stack is now empty
```

Stack dengan queue.LifoQueue

```python
# mendemonstrasi implementasi stack dengan queue. LifoQueue

# menginisialisasi stack
stack = LifoQueue(maxsize=5)

#qsize() untuk menampilkan junlah elemen dalam stack
print(stack.qsize())

#put() untuk memasukkan elemen ke dalam stack
stack.put('m')
stack.put('o')
stack.put('h')
stack.put('a')
stack.put('n')

print("Full:", stack.full())
print("Size:", stack.qsize())

#get() untuk menghapus elemen dari stack dalam urutan LIFO
print('\nElemen di keluarkan dari stack')
print(stack.get())
print(stack.get())
print(stack.get())

print("\nKosong: ", stack.empty())
```

stack dengan single linked list

```python
class Node:
  def __init__(self,value):
    self.value = value
    self.next = None


class Stack:

  # intializing a stack, use a dumy node, witch id aeter for handling adge cases.
  def __init__(self):
    self.head = Node("head")
    self.size = 0

  # String reprensentation of the stack
  def __str__(self):
    cur = self.head.next
    out = ""
    while cur:
      out+= str(cur.value) + "->"
      cur = cur.next
    return out[:-2]

  # Get the current size of the stack
  def getsize(self):
    return self.size

  # check if the stack is empty
  def isEmpty(self):
    return self.size == 0

  # get the top item of the stack
  def peek(self):
    # sanitary check to see if we are peeking an empty stack.
    if self.isEmpty():
      raise Exception("Peeking from an empty stack")
    return self.head.next.value

  # Push a value into the stack
  def push(self,value):
    node = Node(value)
    node.next = self.head.next
    self.head.next = node
    self.size += 1

  # Remove a value from the stack and return
  def pop(self):
    if self.isEmpty():
      raise Exception("Popping from an empty stack")
    remove = self.head.next
    self.head.next = self.head.next.next
    self.size -= 1
    return remove.value


# Driver Code
if __name__ == "__main__":
  stack = Stack()
  for i in range(1, 11):
    stack.push(i)
  print(f"Stack: {stack}")

  for _ in range(1, 6):
    remove = stack.pop()
    print(f"Pop: {remove}")
  print(f"Stack: {stack}")
```
